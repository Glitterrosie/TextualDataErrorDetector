# Weather Dataset
### MinTemp > MaxTemp Transposition

```python
import pandas as pd

data = pd.read_csv('../datasets/weather_subset1_group1_w_errors.csv')
```

```python
def is_numeric(value: str):
    try:
        float(value)
        return True
    except ValueError:
        return False

both_numeric = data[data['MinTemp'].apply(is_numeric) & data['MaxTemp'].apply(is_numeric)]
filtered = both_numeric[both_numeric['MinTemp'].astype(float) > both_numeric['MaxTemp'].astype(float)]
display(filtered)
```

### Rainfall Evaporation Transposition

```python
both_numeric = data[data['Rainfall'].apply(is_numeric) & data['Evaporation'].apply(is_numeric)]
display(both_numeric)
rainfall_contains_153712 = both_numeric[both_numeric['Rainfall'].astype(float) == 15.3712]
# display(rainfall_contains_153712)
```

### Evaporation Sunshine Transposition

```python
both_numeric = data[data['Sunshine'].apply(is_numeric) & data['Evaporation'].apply(is_numeric)]
switched_data = both_numeric[
    (both_numeric['Sunshine'].astype(float) == 15.3712) | (both_numeric['Evaporation'].astype(float) == 14.03)
]
display(switched_data)
```

### Testing the tokenization and spell checking

```python
import re
from spellchecker import SpellChecker

test_value_cast_note = data['cast_note'][0]
print(test_value_cast_note)

spell = SpellChecker()

REGEX = r'\W'
tokens = re.split(REGEX, str(test_value_cast_note))
print(tokens)

for token in tokens:
    if not token:
        continue
    if not spell.known([token]):
        print(f"Misspelled token: {token}")
        suggestions = spell.candidates(token)
        print(f"Suggestions: {suggestions}")
    else:
        print(f"Correctly spelled token: {token}")
```

### Checking dates

```python
KEYBOARD_NEIGHBORS = {
    'a': 'qwsxz',
    'b': 'vghn',
    'c': 'xdfv',
    'd': 'wersfxcv',
    'e': '23wrsdf',
    'f': 'ertdgcv',
    'g': 'rtyfhvb',
    'h': 'tyugjbn',
    'i': '78uojkl',
    'j': 'uihknm',
    'k': 'uiojlm',
    'l': 'iopk',
    'm': 'njk',
    'n': 'bhjm',
    'o': '89ipkl',
    'p': '90ol',
    'q': '1was',
    'r': '34etdfg',
    's': 'qweadzx',
    't': '45ryfgh',
    'u': '67yihjk',
    'v': 'cfgb',
    'w': '12qeasd',
    'x': 'zsdc',
    'y': '56tughj',
    'z': 'asx',
    '0': '9p',
    '1': '2qw',
    '2': '13we',
    '3': '24er',
    '4': '35rt',
    '5': '46ty',
    '6': '57yu',
    '7': '68ui',
    '8': '79io',
    '9': '80op',
}

def is_key_error(word, correct_words_list):
    word_lower = word.lower()
    
    for i, char in enumerate(word_lower):
        neighbors = KEYBOARD_NEIGHBORS.get(char)
        if not neighbors:
            continue

        word_prefix = word_lower[:i]
        word_suffix = word_lower[i+1:]
        for neighbor in neighbors:
            if word_prefix + neighbor + word_suffix in correct_words_list:
                return True
    return False


def label_year(token: str) -> int:
    """
    Distinguish different types of errors in a year token.
    """
    if len(token) == 1: # cases like the decimal part of a number like 1915.w -> token = 'w'
        string_numbers = [str(i) for i in range(0, 10)]
        if is_key_error(token, string_numbers): # we need to check whether this letter is a typo key error, otherwise we label it as OCR
            return "TYPO"
        else:
            return "OCR"

    if len(token) != 4: # cases like 20213 or 203
        return "TYPO"
    
    string_years = [str(i) for i in range(1880, 2025)]
    if is_key_error(token, string_years): # case of exactly 4 numbers
        return "TYPO"

    return "OCR"

print(label_year("a025"))  # OCR
print(label_year("w")) # TYPO
print(label_year("200g")) # OCR
print(label_year("D")) # OCR
print(label_year("Z099.0")) # TYPO
```
