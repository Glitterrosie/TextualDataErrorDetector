```python
import pandas as pd

data = pd.read_csv('../datasets/weather_subset1_group1_w_errors.csv')
# data = pd.read_csv('../datasets/medical_subset1_group1_w_errors.csv')
# data = pd.read_csv('../datasets/imdb_subset1_group1_w_errors.csv')

uniques = {}
for column in data.columns:
    unique_values = data[column].unique()
    uniques[column] = unique_values

max_unique = max(len(values) for values in uniques.values())
unique_frame = pd.DataFrame(columns=data.columns, index=range(max_unique))
for column, values in uniques.items():
    unique_frame[column] = pd.Series(values)

display(unique_frame)

```
